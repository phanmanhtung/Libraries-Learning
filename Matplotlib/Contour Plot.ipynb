```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-2.5,2.5,0.01)
y = np.arange(-2.5,2.5,0.01)
X,Y = np.meshgrid(x,y)
X
```

```python
Y
```

```python
def f(x,y):
    return (1-y**5+x**5)*np.exp(-x**2-y**2)
C = plt.contour(X,Y,f(X,Y),8,colors='black')
plt.contourf(X,Y,f(X,Y),8)
plt.clabel(C,inline=1,fontsize=10)
plt.colorbar()
```

```python
# NOTE: x,y in pair satisfying DOMAIN
x = np.arange(-2.8,2.8,0.01)
y = np.arange(-2.8,2.8,0.01)
X,Y = np.meshgrid(x,y)
def g(x,y):
    return np.sqrt(16-x**2-y**2)
C = plt.contour(X,Y,g(X,Y),8,colors='black')
plt.contourf(X,Y,g(X,Y),8)
plt.clabel(C,inline=1,fontsize=10)
plt.colorbar()
```

```python
x = np.arange(-1.4,1.4,0.01)
y = np.arange(-0.7,0.7,0.01)
X,Y = np.meshgrid(x,y)
def h(x,y):
    return 1/2*np.sqrt(36-9*x**2-36*y**2)
C = plt.contour(X,Y,h(X,Y),8,colors='black')
plt.contourf(X,Y,h(X,Y),8,cmap=plt.cm.hot) # plt.cm.hot -> 'red' type of color
plt.clabel(C,inline=1,fontsize=10)
plt.colorbar()
```

# 3D Surfaces

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-2.5,2.5,0.1)
y = np.arange(-2.5,2.5,0.1)
X,Y = np.meshgrid(x,y)
ax.plot_surface(X,Y,f(X,Y), rstride=1,cstride=1)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,f(X,Y), rstride=1,cstride=1, cmap=plt.cm.hot)
# see from the other side
ax.view_init(elev=30,azim=125)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-2.8,2.8,0.1)
y = np.arange(-2.8,2.8,0.1)
X,Y = np.meshgrid(x,y)
ax.plot_surface(X,Y,g(X,Y), rstride=1,cstride=1, cmap=plt.cm.hot)
```
