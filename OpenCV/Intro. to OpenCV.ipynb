```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Video is measured in FRAMES per second, each FRAME is actually a still image

img = cv2.imread("watch.jpg", cv2.IMREAD_GRAYSCALE) # -> 0
# grayscale - much further simplified, easier to perform analysis on a full color image (RGB - cv2: BGRA w/ A-alpha)
# IMREAD_COLOR -> 1, IMREAD_UNCHANGED -> -1

# Showing the image with cv2
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# save an image
#cv2.imwrite('watchgray.png', img)
```

```python
# Showing the image with matplotlib
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.plot([50,100], [80,100], 'c', linewidth=5)
plt.show()
```

# Interacting with Videos

```python
cap = cv2.VideoCapture(0) # -> 0 means your first webcam in your system

# Want to output the file somewhere
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480)) # in () -> size


while True:
    ret, frame = cap.read() # ret - True or False and the frame
    gray = cv2.cvtColor(frame,  cv2.COLOR_BGR2GRAY) # 'cvt' means 'convert'; '2' means 'to'
    out.write(frame)
    cv2.imshow('frame', frame)
    cv2.imshow('gray', gray)
    
    # get out of the infinite loop (press 'q')
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
# Press 'q' to stop watching
```

# Drawing & Writing on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("watch.jpg", cv2.IMREAD_COLOR)

# DRAW:
cv2.line(img, (0,0), (150,150), (255, 0, 0), 15)
#       image start    end       color:blue linewidth
# color: BGR
# (0, 255, 0) -> green
# (0, 0, 255) -> red

cv2.rectangle(img, (15,25), (200,150), (0, 0, 255), 5)
#                  top-left bottom-right  color  width
# why the color is black??? (NOT cv2.IMREAD_GRAYSCALE) -> cv2.IMREAD_COLOR

# Polygon
pts = np.array([[10,5], [30,5], [49,35], [45,57]], np.int32)
# pts = pts.reshape((-1,1,2)) -> reshape to dot
cv2.polylines(img, [pts], True, (0,255,255), 3)
# True: whether you want the final point to the first point

cv2.circle(img, (100,63), 55, (0, 255, 0), -1)
#                center radius  color   fill-in

# WRITE:
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, 'OpenCV Tuts!', (0,130), font, 1, (200,255,255), 2, cv2.LINE_AA)
#                                starting                       spacing

cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# always press 'q' to dismiss the shown pictures/videos
```
